```python
# Import Libraries
import pandas as pd
import numpy as np
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer
)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# Import Data
df_identification = pd.read_csv('/content/drive/MyDrive/DM2025 - NTU/Homework/data_identification.csv')
df_emotion = pd.read_csv('/content/drive/MyDrive/DM2025 - NTU/Homework/emotion.csv')


import json


path = "/content/drive/MyDrive/DM2025 - NTU/Homework/final_posts.json"

with open(path, 'r') as f:
    raw_json = json.load(f)
```

```python
# Convert json to df

df = pd.DataFrame(raw_json)
df.head()
```

```python
df_flat = pd.json_normalize(df['root'])

df_posts = df_flat.rename(columns={
    '_source.post.post_id': 'post_id',
    '_source.post.text': 'text',
    '_source.post.hashtags': 'hashtags'
})[['post_id', 'text', 'hashtags']]

df_posts.head()

```

```python
df_emotion
```

```python
# Combine json file and identification

df_combined = pd.merge(df_posts, df_identification, left_on='post_id', right_on='id')
df_combined
```

## 2. Feature Engineering Steps



```python
# Data Cleaning - First, I clean the hashtag (extract the word only without "," and "[]") and add them to the text
import ast

def clean_hashtags(x):
    try:
        items = ast.literal_eval(x)
        if isinstance(items, list):
            return " ".join(items)
    except:
        pass
    return ""

df_combined["hashtags_clean"] = df_combined["hashtags"].apply(clean_hashtags)

# Combine
df_combined["input_text"] = (
    df_combined["text"].astype(str) + " " + df_combined["hashtags_clean"]
).str.strip()

df_combined
```

```python
# Text Cleaning - here, me removed URLs and also normalize spaces to make the text cleaner

import re

def clean_text(t):
    t = str(t)
    t = re.sub(r"http\S+|www\S+", "", t)  # remove URLs
    t = re.sub(r"\s+", " ", t).strip()    # normalize spaces
    return t

df_combined["input_text"] = df_combined["input_text"].apply(clean_text)
df_combined
```

```python
# Here I dropped unnecessary columns

df_combined = df_combined.drop(columns=[
    "text",
    "hashtags",
    "id",
    "hashtags_clean"
])
df_combined
```

```python
df_combined
```

```python
# Split Train and Test
df_train = df_combined[df_combined["split"] == "train"]
df_test = df_combined[df_combined["split"] == "test"]
```

```python
# Join df_train and df_emotion
df_train = df_train.merge(df_emotion, left_on="post_id", right_on="id")
df_train

```

```python
# Label Encoding for df_train
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_train["label"] = le.fit_transform(df_train["emotion"])
```

```python
df_train
```

```python
# Drop unnecessary columns
df_train = df_train.drop(columns=["split", "id"])
```

```python
df_test
```

```python
df_test = df_test.drop(columns=["split"])
```

## 3. Model Implementation Steps

```python
# Convert into HuggingFace Dataset

from datasets import Dataset

train_ds = Dataset.from_pandas(df_train[["input_text", "label"]])
test_ds  = Dataset.from_pandas(df_test[["input_text"]])

```

```python
train_ds
```

```python
test_ds
```

```python
# Here, I am using the Twitter Roberta Base Emotion
model_name = "cardiffnlp/twitter-roberta-base-emotion"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=df_train["label"].nunique(),
    ignore_mismatched_sizes=True
)
```

```python
def tokenize(batch):
    return tokenizer(
        batch["input_text"],
        truncation=True,
        padding="max_length",
        max_length=196
    )

```

```python
train_tok = train_ds.map(tokenize, batched=True)
test_tok  = test_ds.map(tokenize, batched=True)

train_tok = train_tok.remove_columns(["input_text"])
test_tok  = test_tok.remove_columns(["input_text"])

train_tok.set_format("torch")
test_tok.set_format("torch")


```

```python
training_args = TrainingArguments(
    output_dir="/twitter_emotion_model",
    save_strategy="no",
    logging_steps=200,
    per_device_train_batch_size=16,
    num_train_epochs=3,               # I chose 3 to make it balanced between the resources needed and the performance it gets
    learning_rate=3e-5,
    weight_decay=0.01,
)

```

```python
from transformers import Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_tok,
    eval_dataset=test_tok,
    tokenizer=tokenizer,
)
```

```python
trainer.train()
```

#### Predicting

```python
import numpy as np

test_logits = trainer.predict(test_tok).predictions
test_pred_ids = np.argmax(test_logits, axis=1)
```

```python
# Inverting back to emotion words
test_pred_labels = le.inverse_transform(test_pred_ids)
```

```python
# Creating a file to submit on Kaggle
submission = pd.DataFrame({
    "id": df_test["post_id"],
    "emotion": test_pred_labels
})
```

### 2.2 Insights Gained
